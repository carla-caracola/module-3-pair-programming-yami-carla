```python
# Ejercicio 1: ¿Cómo se distribuye la tasa de natalidad en el conjunto de datos? ¿Hay algún patrón evidente?

sns.histplot(x="birth_rate", data=df, kde=True, bins=20);
```

```python
sns.boxplot(x="birth_rate", data=df);
```

```python
# Conclusiones del ejercicio 1: 
    # La tasa de natalidad de la mayoría de los países tiene un valor entre 10 y 30. 
```

```python
# Ejercicio 2: ¿Existe una relación aparente entre la tasa de mortalidad infantil y la esperanza de vida? 
# Proporciona una breve descripción de los patrones que observes.
```

```python
sns.regplot(x="infant_mortality", y="life_expectancy", data=df);
```

```python
# conclusiones del ejercicio 2: si, se observa una relación estrecha entre ambas variables, siendo la misma inversamente proporcional, es decir que, a menor tasa de mortalidad infantil, mayor es la esperanza de vida. 
```

```python
# Ejercicio 3: ¿Cuál es el continente con la mayor cantidad de países en el conjunto de datos? Proporciona el porcentaje de países en ese continente.

# creamos un nuevo dataframe agrupando los datos por continente

df_grouped = df.groupby("continent")["country"].count().reset_index()

df_grouped
```

```python
# creamos un pie chart

plt.pie ("country", labels="continent", data= df_grouped, autopct="%1.1f%%", textprops={"color":"white"});
```

```python
# conclusiones ejercicio 3: el país con mayor cantidad de países en el conjunto de datos es África con un 27.3 % 
```
