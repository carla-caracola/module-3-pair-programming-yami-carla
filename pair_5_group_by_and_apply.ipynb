```python
# 1) Las columnas land_area, armed_forces_size, co2-emissions, gdp, population, urban_population, latitude, 
# longitude, density, cpi son de tipo object pero en realidad deberían ser numéricas. 
# Crea una función que nos permita cambiar el tipo de dato de estas columnas al que le corresponde.

def df_cambiar_coma(cadena):
    try:
        return float(cadena.replace(",","."))
    except: 
        return np.nan

list_to_transform = ["population","land_area", "armed_forces_size", "co2-emissions", "gdp", "population", "urban_population", "latitude", "longitude", "density", "cpi"]

for col in list_to_transform:
    df[col] = df[col].apply(df_cambiar_coma)
    print(df[col].dtype) 
```
